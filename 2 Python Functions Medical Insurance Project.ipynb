12. Congratulations! You have successfully created a useful function that calculates the “medical insurance costs” based on the values passed into the function.

    - Create a second function to calculate the difference between the insurance costs (given as inputs) of any two individuals, then print a statement saying:<br>
    `"The difference in insurance cost is xxx dolars."`
    


```python
# Create calculate_insurance_cost() function below: 
def calculate_insurance_cost(name, age, sex, bmi, num_of_children, smoker):
  estimated_cost = 250*age - 128*sex + 370*bmi + 425*num_of_children + 24000*smoker - 12500
  output_message = f"The estimated insurance cost for {name} is {estimated_cost} dollars."
  return estimated_cost, output_message

# Create diff_btwn_insurance_cost() function below:
def diff_btwn_insurance_cost(name1, name2, cost1, cost2):
  cost_difference = cost1 - cost2
  if cost_difference < 0:
    print(f"{name1}'s insurance cost is less than {name2}'s insurance cost by {abs(cost_difference)} dollars.")
  elif cost_difference > 0:
    print(f"{name1}'s insurance cost is greater than {name2}'s insurance cost by {abs(cost_difference)} dollars.")
  else:
    print(f"{name1} and {name2} have the same insurance cost.")

# Estimate Maria's insurance cost
maria_insurance_cost, maria_message = calculate_insurance_cost(name = "Maria", age = 28, sex = 0, bmi = 26.2, num_of_children = 3, smoker = 0)
print(maria_message)

# Estimate Omar's insurance cost 
omar_insurance_cost, omar_message = calculate_insurance_cost(name = "Omar", age = 35, sex = 1, bmi = 22.2, num_of_children = 0, smoker = 1)
print(omar_message)

# Estimate Auggi's insurance cost 
auggi_insurance_cost, auggi_message = calculate_insurance_cost(name = "Auggi", age = 29, sex = 1, bmi = 26.5, num_of_children = 0, smoker = 0)
print(auggi_message)

# Estimate the difference between two individuals' insurance costs (e.g. Maria and Auggi)
diff_btwn_insurance_cost("Maria", "Auggi", maria_insurance_cost, auggi_insurance_cost)

# Estimate the difference between two individuals' insurance costs (e.g. Maria and Omar)
diff_btwn_insurance_cost("Maria", "Omar", maria_insurance_cost, omar_insurance_cost)
```
